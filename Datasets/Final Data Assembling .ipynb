```python
#Import the pandas library.
import pandas as pd 
```

```python
#Loading the transport dataset and fuel price dataset.
transport_df = pd.read_csv("Transport Data.csv")
fuel_df = pd.read_csv("fuel_prices.csv")
```

```python
#Step 1 - Data Preprocessing (Conversion of Date Column)
transport_df['Date'] = pd.to_datetime(transport_df['Date'])
fuel_df['Date'] = pd.to_datetime(fuel_df['Date'])
```

```python
#Step 2 - Data Merging (Combine two dataframes into a new dataframe 'final_df')
final_df = pd.merge(
    transport_df, 
    fuel_df, 
    on=['Date', 'City'], #Merge rows where both values match.
    how='left' #keeps all rows from the left dataframe.
)
```

```python
#Step 3 - Data Inspection (Checks values for missing values NaN).
final_df.isna().sum().sum()
```

```python
#Step 4 - Data Cleaning 
final_df = final_df.drop(['Petrol_Price_y','Diesel_Price_y'], axis=1)
```

```python
final_df.rename(columns={'Petrol_Price_x': 'Petrol_Price'}, inplace=True)
final_df.rename(columns={'Diesel_Price_x': 'Diesel_Price'}, inplace=True)
```
